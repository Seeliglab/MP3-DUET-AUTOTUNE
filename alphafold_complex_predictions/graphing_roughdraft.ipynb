```python
#plot performance regression 
#r2
#spearman 
def plot_df_regression_padj_final(df, savename, w =4, h =2 ):
    fig, ax = plt.subplots(2,1, sharex= True) #plot AUCROC, MCC, n_features
    # the size of A4 paper
    fig.set_size_inches(w,h)
    
    plt.xlabel('')
    #plotting baseline AUCROC for imbalanced class +
    s =sns.barplot(data = df,
                x = 'dataset', 
                y = 'test_r2', ax = ax[0])
    #s._legend.remove()
    ax[0].set_ylim(0,1)
    plt.legend([],[], frameon=False)
    plt.xlabel('')
    s = sns.barplot(data = df,
                x = 'dataset', 
                y = 'test_spearman', ax = ax[1])
    #s._legend.remove()
    ax[1].set_ylim(0,1)
    
    plt.xticks(rotation=90)
    plt.legend([],[], frameon=False)
    plt.xlabel('')
    for i in range(0,2):
        ax[i].legend([],[], frameon=False)
        ax[i].set_xlabel('')

    plt.savefig(savename, dpi = 300)
    plt.show()
    

def plot_df_classification_padj_final(df, savename, w =4, h =2 ):
    fig, ax = plt.subplots(2,1, sharex= True) #plot AUCROC, MCC, n_features
    # the size of A4 paper
    fig.set_size_inches(w,h)
    
    plt.xlabel('')
    #plotting baseline AUCROC for imbalanced class +
    s =sns.barplot(data = df,
                x = 'dataset', 
                y = 'test_AUCROC', ax = ax[0])
    #s._legend.remove()
    ax[0].set_ylim(0,1)
    ax[0].axhline(0.5)
    s = sns.barplot(data = df,
                x = 'dataset', 
                y = 'test_avgpr', ax = ax[1])
    #s._legend.remove()
    ax[1].set_ylim(0,1)
    ax[1].axhline(overall_baseline_avgpr)
    plt.xticks(rotation=90)
    for i in range(0,2):
        ax[i].legend([],[], frameon=False)
        ax[i].set_xlabel('')

    plt.savefig(savename, dpi = 300)
    plt.show()
```
