```python
#make enrichment versions 

vals_counts = ['trp1_new_1_smaller', 'his1_new_1_smaller', 'trp1_new_2_smaller', 'his1_new_2_smaller']
for vc in vals_counts:
    deseq_values.loc[deseq_values[vc]  == 0, vc] = None

deseq_values['lin_enrich'] = calc_lin_enrichment(deseq_values['trp1_new_1_smaller'], deseq_values['his1_new_1_smaller'])
deseq_values['log2_enrich_1'] = calc_log2_enrichment(deseq_values['lin_enrich'])
deseq_values.loc[(deseq_values.trp1_new_1_smaller.isna()) & (deseq_values.his1_new_1_smaller.isna()), 'lin_enrich'] = None
deseq_values.loc[(deseq_values.trp1_new_1_smaller.isna()) & (deseq_values.his1_new_1_smaller.isna()), 'log2_enrich_1'] = None

deseq_values['lin_enrich_2'] = calc_lin_enrichment(deseq_values['trp1_new_2_smaller'], deseq_values['his1_new_2_smaller'])
deseq_values['log2_enrich_2'] = calc_log2_enrichment(deseq_values['lin_enrich_2'])
deseq_values.loc[(deseq_values.trp1_new_2_smaller.isna()) & (deseq_values.his1_new_2_smaller.isna()), 'lin_enrich_2'] = None
deseq_values.loc[(deseq_values.trp1_new_2_smaller.isna()) & (deseq_values.his1_new_2_smaller.isna()), 'log2_enrich_2'] = None
```
