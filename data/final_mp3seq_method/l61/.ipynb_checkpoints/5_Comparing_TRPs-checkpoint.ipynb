## TRP 60h

```python
trp_60h_df = pd.read_csv('L61_TRP_60h_table.csv', index_col=0)
```

```python
trp_60h_df.shape
```

```python
trp_60h_df
```

```python


v_24, v_48, v_60, v_24_log, v_48_log, v_60_log = [], [], [], [], [], []

for binder1 in trp_24h_df.columns:
    for binder2 in trp_24h_df.index:
        
        v_24.append(trp_24h_df[binder1][binder2])
        v_48.append(trp_48h_df[binder1][binder2])
        v_60.append(trp_60h_df[binder1][binder2])
        
        v_24_log.append(math.log(trp_24h_df[binder1][binder2]))
        v_48_log.append(math.log(trp_48h_df[binder1][binder2]))
        v_60_log.append(math.log(trp_60h_df[binder1][binder2]))

        
```

```python
stats.pearsonr(v_24, v_48)[0]**2
```

```python
stats.pearsonr(v_48, v_60)[0]**2
```

```python
stats.pearsonr(v_24_log, v_48_log)[0]**2
```

```python
stats.pearsonr(v_48_log, v_60_log)[0]**2
```

```python

plt.figure(figsize=(5.0, 5.0), dpi=100)

#plt.xlim([50, 5000])
#plt.ylim([50, 5000])

plt.xlabel("L61 24 hours (LINEAR)", size=16)
plt.ylabel("L61 48 hours (LINEAR)", size=16)

plt.scatter(v_24, v_48)
```

```python


plt.figure(figsize=(5.0, 5.0), dpi=100)

#plt.xlim([50, 5000])
#plt.ylim([50, 5000])

plt.xlabel("L61 48 hours (LINEAR)", size=16)
plt.ylabel("L61 60 hours (LINEAR)", size=16)

plt.scatter(v_48, v_60)
```

```python

plt.figure(figsize=(5.0, 5.0), dpi=100)

#plt.xlim([50, 5000])
#plt.ylim([50, 5000])

plt.xlabel("L61 24 hours (LOG)", size=16)
plt.ylabel("L61 48 hours (LOG)", size=16)

plt.scatter(v_24_log, v_48_log)
```
