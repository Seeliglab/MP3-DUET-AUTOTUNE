```python


plt.figure(figsize=(6.5, 5.4), dpi=100)


result = sns.heatmap(lin_2mM_3AT_df_trunc, cmap=cmap, vmax=10, cbar_kws={'label': 'HT-Y2H Linear Enrichment Score'})

ax = result.axes

ax.figure.axes[-1].yaxis.label.set_size(14)

plt.xlabel('Activation domain', size = 14)
plt.ylabel('Binding domain', size = 14)

for i in range(int(0.5*len(trp_df.loc[new_columns_order][new_columns_order]))):
    rect = patches.Rectangle((1+2*i,2*i),1,1,linewidth=2,edgecolor='yellow',facecolor='none')
    ax.add_patch(rect)
    rect = patches.Rectangle((2*i,1+2*i),1,1,linewidth=2,edgecolor='yellow',facecolor='none')
    ax.add_patch(rect)

```

```python


v_up, v_up_log, v_low, v_low_log = [], [], [], []

threshold = 20


for i in range(trp_df.shape[0]):
    for j in range(i+1, trp_df.shape[0]):
        
        trp_count_up  = trp_df.values[i][j]
        trp_count_low = trp_df.values[j][i]
        
        val_up  = his_2mM_3AT_df.values[i][j]
        val_low = his_2mM_3AT_df.values[j][i]

        if trp_count_up > threshold and trp_count_low > threshold and ~np.isnan(val_up) and ~np.isnan(val_low):

            v_up.append(val_up/trp_count_up)
            v_low.append(val_low/trp_count_low)

            v_up_log.append(math.log(val_up/trp_count_up))
            v_low_log.append(math.log(val_low/trp_count_low))

            
```

```python
len(v_up)
```

### Flask_1 up and low correlation

```python
stats.pearsonr(v_up, v_low)[0]**2
```

```python
stats.pearsonr(v_up_log, v_low_log)[0]**2
```

```python
plt.figure(figsize=(5.0, 5.0), dpi=100)

#plt.xlim([50, 5000])
#plt.ylim([50, 5000])

plt.xlabel("Upper triangle", size=16)
plt.ylabel("Lower triangle", size=16)

plt.scatter(v_up, v_low)
```

```python
plt.figure(figsize=(5.0, 5.0), dpi=100)

#plt.xlim([50, 5000])
#plt.ylim([50, 5000])

plt.xlabel("Upper triangle", size=16)
plt.ylabel("Lower triangle", size=16)

plt.scatter(v_up_log, v_low_log)
```

# Comparing V3 HIS counts

```python

threshold = 1

v_2mM_3AT_lin, v_10mM_3AT_lin, v_20mM_3AT_lin = [], [], []
v_2mM_3AT_log, v_10mM_3AT_log, v_20mM_3AT_log = [], [], []

for binder1 in trp_df.columns:
    for binder2 in trp_df.index:
        
        if his_2mM_3AT_df[binder1][binder2] > threshold and his_10mM_3AT_df[binder1][binder2] > threshold and \
           his_20mM_3AT_df[binder1][binder2] > threshold:
            
            v_2mM_3AT_lin.append(his_2mM_3AT_df[binder1][binder2])
            v_10mM_3AT_lin.append(his_10mM_3AT_df[binder1][binder2])
            v_20mM_3AT_lin.append(his_20mM_3AT_df[binder1][binder2])

            v_2mM_3AT_log.append(math.log(his_2mM_3AT_df[binder1][binder2]))
            v_10mM_3AT_log.append(math.log(his_10mM_3AT_df[binder1][binder2]))
            v_20mM_3AT_log.append(math.log(his_20mM_3AT_df[binder1][binder2]))
            
```
