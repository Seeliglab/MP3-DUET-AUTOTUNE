```python
L67_binders_df["Length"] = L67_binders_df["N_Term_Trunc_Binder_Seq"].apply(lambda x: len(x))
```

```python
import pandas as pd
from Bio import SeqIO
from Bio.Seq import Seq
```

```python
L67_binders_df['translation'] = L67_binders_df.N_Term_Trunc_Binder_Seq.apply(lambda x: str(Seq(x).translate()))
```

```python

L67_short_binders_list = []

L = 140 

for i in range(int(L67_binders_df.shape[0]/2)):
    
    binder_A_len = L67_binders_df.loc[2*i].Length
    binder_B_len = L67_binders_df.loc[2*i+1].Length

    if max(binder_A_len, binder_B_len) <= L:
        
        L67_short_binders_list.append(L67_binders_df.loc[2*i].Binder_Name)
        L67_short_binders_list.append(L67_binders_df.loc[2*i+1].Binder_Name)
        
        print(L67_binders_df.loc[2*i].Binder_Name)
        print(L67_binders_df.loc[2*i].N_Term_Trunc_Binder_Seq)
        print(L67_binders_df.loc[2*i].Length)
        print()

        print(L67_binders_df.loc[2*i+1].Binder_Name)
        print(L67_binders_df.loc[2*i+1].N_Term_Trunc_Binder_Seq)
        print(L67_binders_df.loc[2*i+1].Length)
        print()
```
