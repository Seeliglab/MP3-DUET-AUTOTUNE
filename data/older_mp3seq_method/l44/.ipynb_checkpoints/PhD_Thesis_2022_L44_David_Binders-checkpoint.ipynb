### Correlations

```python


v44_main_lin, v44_main_log, v44_reciprocal_lin, v44_reciprocal_log = [], [], [], []

v_stephanie_lin, v_stephanie_log = [], []


for binder_x in davids_binders_x:
    
    for binder_y in davids_binders_y:

        v_stephanie_lin.append(stephanie_df[binder_x][binder_y])
        v_stephanie_log.append(-math.log(stephanie_df[binder_x][binder_y]))

        v44_main_lin.append(L44_lin_df[binder_x][binder_y])
        v44_main_log.append(L44_log_df[binder_x][binder_y])
        
        v44_reciprocal_lin.append(L44_lin_df[binder_y][binder_x])
        v44_reciprocal_log.append(L44_log_df[binder_y][binder_x])

        
```

```python
len(v44_main_lin)
```

```python
stats.spearmanr(v44_main_log, v_stephanie_log)[0]
```

```python
stats.spearmanr(v44_reciprocal_log, v_stephanie_log)[0]
```

```python
stats.pearsonr(v44_main_log, v_stephanie_log)[0]**2
```

```python
stats.pearsonr(v44_reciprocal_log, v_stephanie_log)[0]**2
```
