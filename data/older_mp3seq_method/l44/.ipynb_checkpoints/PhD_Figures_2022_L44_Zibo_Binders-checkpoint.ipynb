```python
import os
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
from matplotlib import gridspec
import seaborn as sns
import math
import matplotlib.patches as patches
```

```python
cmap = sns.cubehelix_palette(light=1, as_cmap=True)
```

### L44 TRP table

```python
L44_trp_df = pd.read_csv("L44_TRP_sorted_table.csv", index_col=0)
```

```python
L44_trp_df.shape
```

```python
L44_trp_df.fillna(0).values.sum()
```

```python
zibo_binders = ["A_ZCON_13", "B_ZCON_13",
                "A_ZCON_150", "B_ZCON_150",
                "A_ZCON_154", "B_ZCON_154",
                "A_ZCON_155", "B_ZCON_155",
                "A_ZCON_15_re_loop", "B_ZCON_15_re_loop",
                "A_ZCON_162", "B_ZCON_162",
                "A_ZCON_37_3", "B_ZCON_37_3",
                "A_ZCON_37", "B_ZCON_37",
                "A_ZCON_6", "B_ZCON_6"]
```

```python
L44_trp_df.loc[zibo_binders][zibo_binders]
```

```python

plt.figure(figsize=(5, 4.0), dpi=100)

result = sns.heatmap(L44_trp_df.loc[zibo_binders][zibo_binders], cmap = cmap, vmax=400)

ax = result.axes

for i in range(int(0.5*len(L44_trp_df.loc[zibo_binders][zibo_binders]))):
    rect = patches.Rectangle((1+2*i,2*i),1,1,linewidth=2,edgecolor='yellow',facecolor='none')
    ax.add_patch(rect)
    rect = patches.Rectangle((2*i,1+2*i),1,1,linewidth=2,edgecolor='yellow',facecolor='none')
    ax.add_patch(rect)

plt.show()
```
